```python
import fitz
print(fitz.__doc__)
```

```python
filepath = 'pdf/285998.pdf'

doc = fitz.open(filepath)
```

```python
doc.pageCount
```

```python
doc.metadata
```

```python
doc.getToC()
```

```python
page = doc.loadPage(0)  # loads page number 'pno' of the document (0-based)
# get all links on a page
links = page.getLinks()
print(f'Links: {len(links)}')
for link in links:
    print(link)
    
pix = page.getPixmap()
pix.writeImage("pdf/page-%i.png" % page.number)

text = page.getText("text")
print(text)
```
