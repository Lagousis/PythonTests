Simple tutorial

```python
def functionToFind(x1, x2, x3):
    #y = 3 * x1**2 - 7 * x2 + 13/(x3+1)
    y = 3 * x1 - 7 * x2 + 13*x3
    return y
```

```python
print(functionToFind(5, 1, 4))
```

```python
# Create random data
number_of_samples = 1000
import random
import numpy as np
x = list(zip(random.sample(range(1000), number_of_samples), random.sample(range(1000), number_of_samples), random.sample(range(1000), number_of_samples)))
y=[]
for input in x:
    y.append(functionToFind(input[0], input[1], input[2]))
y=np.array(y)
len(x), len(y)
x,y
```

```python
#Normalization
#from sklearn import preprocessing
#normalized_x = preprocessing.normalize(x, axis=1)

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

scaler_x = StandardScaler()
scaled_x = scaler_x.fit_transform(x)

scaler_y = StandardScaler()
scaled_y = scaler_y.fit_transform(y.reshape(-1,1))

scaled_x, scaled_y
```

```python
# y.reshape(-1,1) converts 1D array to 2D
# One shape dimension can be -1. In this case, the value is inferred from the length of the array and remaining dimensions.
y.shape, y.reshape(-1,1).shape
```

```python
import torch
from torch import nn
from torch import optim
import torch.nn.functional as F
```

Convert numpy arrays to tensors

```python
x_train, y_train = map(torch.tensor, (scaled_x, scaled_y))
#x_train, y_train
print(x_train.dim(), x_train.type())
x_train = x_train.type(torch.FloatTensor)
y_train = y_train.type(torch.FloatTensor)
print(x_train.dim(), x_train.type(), len(x_train))
```

Create NN

```python
class myModel(nn.Module):
    def __init__(self):
        #super().__init__()
        super(myModel, self).__init__() 
        # Calling Super Class's constructor
        self.lin1 = nn.Linear(3, 10)
        self.lin2 = nn.Linear(10,1)

    def forward(self, xb):
        x = F.relu(self.lin1(xb))
        return self.lin2(x)
```

```python
def get_model(lr):
    model = myModel()
    return model, optim.SGD(model.parameters(), lr=lr)
```

```python
loss_func = F.mse_loss #nn.MSELoss()# Mean Squared Loss
```

```python
model, opt = get_model(1e-1)
print(loss_func(model(x_train), y_train))
```

Train without loader

```python
model, opt = get_model(1e-1)
print(f'Parameters: {model.parameters}')

n = len(x_train)
bs = 100
epochs = 7000

for epoch in range(epochs):
    for i in range((n - 1) // bs + 1):
        start_i = i * bs
        end_i = start_i + bs
        xb = x_train[start_i:end_i]
        yb = y_train[start_i:end_i]
        pred = model(xb)
        loss = loss_func(pred, yb)
        #if i==0:
        #    print(xb, pred, yb)        
        
        loss.backward()
        opt.step()
        opt.zero_grad()

    if (epoch % 100) == 0:
        print('epoch {}, loss {}'.format(epoch,loss))
            
#print(loss_func(model(xb), yb))
print(loss_func(model(x_train), y_train))
```

Train with validation and fit

```python
def loss_batch(model, loss_func, xb, yb, opt=None):
    loss = loss_func(model(xb), yb)

    if opt is not None:
        loss.backward()
        opt.step()
        opt.zero_grad()

    return loss.item(), len(xb)

import numpy as np

# Note that we always call model.train() before training, and model.eval() before inference, 
# because these are used by layers such as nn.BatchNorm2d and nn.Dropout to ensure appropriate behaviour for these 
# different phases.

def fit(epochs, model, loss_func, opt, train_dl, valid_dl):
    for epoch in range(epochs):
        model.train()
        for xb, yb in train_dl:
            loss_batch(model, loss_func, xb, yb, opt)

        if (epoch % 100) == 0:
            model.eval()
            with torch.no_grad():
                # get the loss and number of samples for each batch in valid_dl
                losses, nums = zip(
                    *[loss_batch(model, loss_func, xb, yb) for xb, yb in valid_dl]
                )
            val_loss = np.sum(np.multiply(losses, nums)) / np.sum(nums)

            print(epoch, val_loss)

from torch.utils.data import DataLoader
        
def get_data(train_ds, valid_ds, bs):
    return (
        DataLoader(train_ds, batch_size=bs, shuffle=True),
        DataLoader(valid_ds, batch_size=bs * 2),
    )
```

```python
from torch.utils.data import TensorDataset

train_ds = TensorDataset(x_train, y_train)
# Use the train dataset as validation, just for test not a good practice
valid_ds = TensorDataset(x_train, y_train)

train_dl, valid_dl = get_data(train_ds, valid_ds, bs)
model, opt = get_model(1e-1)
fit(epochs, model, loss_func, opt, train_dl, valid_dl)
```
