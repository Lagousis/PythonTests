```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.dates import DateFormatter
from matplotlib.dates import DayLocator
```

```python
df = pd.read_excel('data/PobucaNewUsersAthens.xlsx')
df.head(10)
```

```python
df['Date'] = pd.to_datetime(df['Signed up (EET)']).apply(lambda x: x.date())
```

```python
grouped = df.groupby(['Date']).count()['Name'].reset_index()
```

```python
fig, ax = plt.subplots()
grouped.plot.line(x='Date', y='Name', ax=ax, figsize=(20, 5))
fig.autofmt_xdate()

ax.set(xlabel="Date",
       ylabel="Registrations",
       title="Registrations per date")

#set ticks every day
ax.xaxis.set_major_locator(DayLocator())

ax.xaxis.set_major_formatter(DateFormatter("%m/%d"))
ax.get_legend().remove()

#ax.set_xticklabels(df['Date'])
#plt.xticks(rotation=45)
```
